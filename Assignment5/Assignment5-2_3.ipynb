### Question 1 - Part A (XML)

#### For the XML file get a list of all data attributes and print them. Then print the first 5 calEvent Nodes

```python
import xml.etree.ElementTree as ET
from pprint import pprint

attributes = []
calEvents = []

#Prompt the user to enter file name
fileName = input("Enter the file name: ")
#Parse an xml file by name
tree = ET.parse(fileName)
root = tree.getroot()

#Prompt the user to enter tag name
tagName = input("Enter the path to tag (eg: viewentry/entrydata): ")
    
#Search for attributes
for child in root.findall(tagName):         
    attribute = child.attrib
    #Remove duplicates - attributes
    if attribute not in attributes:
        attributes.append(attribute)
    #for calEvent in child:
     #   if (calEvent not in calEvents) and (calEvent.text != None):
      #      calEvents.append(calEvent.text)
            
#Search for calEvent nodes           
for calEvent in root.findall(".//entrydata[@name='EventName']/text"):
    calEvents.append(calEvent.text)
    
#Print attributes of entered tag
print("\nAll " + tagName + " attributes: ")
for val in attributes:
    pprint(val)
    
#Print top 5 calEvents
print("\nTop 5 calEvents: ")
for val in calEvents[:5]:
    pprint(val)
```

### Question 1 - Part B (JSON)

#### For the JSON file get a list of Keys and print them. Then print the last 5 JSON calEvent Objects 

```python
import json
from pprint import pprint

keys = []

fileName = input("Enter the file name: ")
fhand=open(fileName)
jsonstr = fhand.read()
jdata = json.loads(jsonstr)

length = len(jdata)

for x in range(0,length):
    #Parsing through first level of dictionaries
    for key,value in jdata[x].items():
        if key not in keys:
            keys.append(key)
        for key,value in value.items():
            if key not in keys:
                keys.append(key)
                #Parsing through second level of dictionaries
                if type(value) == dict:
                    for key,value in value.items():
                        if key not in keys:
                            keys.append(key)
                #Parsing through list
                if type(value) == list:
                    for x in range(0,len(value)):
                        for key in value[x]:
                            if key not in keys:
                                keys.append(key)
                
#Print all keys
for key in keys:
    pprint(key)
    
#Print the last 5 calEvent objects
for key in jdata[-5:]:
    pprint("Key:")
    pprint(key)
```
